## Question 2

````python
def odd_even(x):
    """ Function to check whether the integer is odd or even
    Input
    ---
    x: must be integer

    Return
    ---
    a statement of wheter the integer is odd or even"""
    if isinstance(x, int):
        if (x % 2 == 0) and (x != 0):
            print(f"{x} is even")
        elif (x % 2 == 1):
            print(f"{x} is odd")
        elif x == 0:
            print(f"{x} is neither odd nor even")

    else:
        print(f"{x} is not an integer")
````

````python
odd_even(5.6)
odd_even(6)
odd_even(1)
odd_even(0)
odd_even(-34)
````

## Question 3
Write a simple example code to show that Numpy is more efficient in numerical computation of large arrays of data than equivalent Python list.

````python
import time

# Create a large array using NumPy
n = 10000000
numpy_array = np.ones(n)

startTime = time.time()
# Compute the sum using NumPy
numpy_sum = np.sum(numpy_array)
endTime = time.time()

# Print time consumed
print(f"Sum using NumPy: {numpy_sum}")
print(f"The Elapsed time for NumPy is {endTime - startTime}")

# Now do the same with a Python list
python_list = [1] * n

startTime = time.time()
# Compute the sum using Python list
python_sum = sum(python_list)
endTime = time.time()

# Print time consumed
print(f"Sum using Python list: {python_sum}")
print(f"The Elapsed time for Python list is {endTime - startTime}")

# Check which method more efficient
print(f"\nResult:")
if numpy_sum > python_sum:
  print(f"Python list is more efficient")
else:
  print(f"NumPy is more efficient")
````

## Question 5

````python
# 256x256, 3 colour channels (R, G, B)
canvas = np.zeros((256, 256, 3), dtype=np.uint8)

# Define circle parameters
center_x, center_y = 128, 128
radius = 50
color = (255, 255, 255)  # White color

# Draw the circle
for x in range(256):
    for y in range(256):
        # Equation of Circle: x^2 + y^2 = r^2
        if (x - center_x) ** 2 + (y - center_y) ** 2 <= radius ** 2:
            canvas[y, x] = color

# Display the image
plt.imshow(canvas)
plt.axis('off')
plt.show()
````
